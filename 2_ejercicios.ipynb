```python
import numpy as np
```

```python
# ejercicio 1
# se usó el np.all para ver si todos los valores son True
arreglo_cero= (np.random.randint(0, 10, size=[10,10]))
print(arreglo_cero)
indices_ceros = np.all(arreglo_cero != 0)
print ("¿El resultado no tiene ceros ? ", indices_ceros)
        
```

```python
# ejercicio 2 Dado el arreglo `a = np.array([0, 0, 2, 0])`, determina si contiene algún elemento que no sea cero.
# np.any es para ver si en algun elemento es diferente a 0
arreglo_cero= (np.random.randint(0, 10, size=[10,10]))
print(arreglo_cero)
indices_ceros = np.any(arreglo_cero !=0)
print ("¿El resultado contiene algún elemento que no sea cero ? ", indices_ceros)
        
```

```python
# ejercicio 3 Analiza el arreglo `a = np.array([1, 2, np.nan, np.inf])` y determina si contiene valores que sean infinitos o valores no numéricos (NaN).
arreglo_cero= (np.random.randint(0, 10, size=[5,5]))
print(arreglo_cero)
arreglo_cero = np.array([1, 2, np.nan, np.inf])
infinito= arreglo_cero = np.inf  # infinito
print(infinito)
valor_no_numerico= np.nan  # no es un número
print(valor_no_numerico)
```

```python
#Ejercicio 4. Dado el arreglo `a = np.array([1, 2, np.inf])`, determina si contiene valores infinitos.
a = np.array([1, 2, np.inf])
infinito = np.isinf(a)

print(infinito)
print("¿El arreglo contiene infinitos?:", np.any(infinito))
```

```python
# Ejercicio 5. Dado el arreglo `a = np.array([1, np.nan, 3])`, determina si contiene valores NaN.
a = np.array([1, np.nan, 3])

tiene_nan = np.isnan(a)

print(tiene_nan)
print("¿El arreglo contiene NaN?:", np.any(tiene_nan))
```

```python
#Ejercicio 6. Dados los arreglos `a = np.array([1, 3, 5])` y `b = np.array([2, 3, 4])`, realiza una comparación elemento a elemento contemplando mayor, mayor o igual, menor y menor o igual.
a = np.array([1, 3, 5])
b = np.array([2, 3, 4])

mayor = a > b
mayor_igual = a >= b
menor = a < b
menor_igual = a <= b

print("a > b:", mayor)
print("a >= b:", mayor_igual)
print("a < b:", menor)
print("a <= b:", menor_igual)
```

```python
#Ejercicio 7. Crea un arreglo que contenga los valores 1, 7, 13 y 105, y determina la cantidad de memoria que ocupa el arreglo.

a = np.array([1, 7, 13, 105])

memoria = a.nbytes

print("Arreglo:", a)
print("Memoria ocupada:", memoria, "bytes")
```

```python
#Ejercicio 8. Crea un arreglo de 10 ceros; uno de 10 unos; y uno de 10 cincos.

ceros = np.zeros(10)


unos = np.ones(10)


cincos = np.full(10, 5)

print("Ceros:", ceros)
print("Unos:", unos)
print("Cincos:", cincos)
```

```python
#Ejercicio 9. Crea un arreglo que contenga todos los números enteros entre 30 y 70.
a = np.arange(30, 71)

print(a)
```

```python
#Ejercicio 10. Crea un arreglo que contenga los números pares entre 30 y 70.
a = np.arange(30, 71, 2)

print(a)
```

```python
#Ejercicio 11. Crea una matriz identidad de 3x3.

I = np.eye(3)

print(I)
```

```python
#Ejercicio 12. Genera un número aleatorio entre 0 y 1.

num = np.random.rand()

print(num)
```

```python
#Ejercicio 13. Crea un arreglo de 15 números aleatorios tomados de una distribución normal estándar.
a = np.random.randn(15)

print(a)
```

```python
#Ejercicio 14. Crea un arreglo con valores entre 15 y 55, e imprime todos los valores excepto el primero y el último.
a = np.arange(15, 56)   

print(a[1:-1])
```

```python
#Ejercicio 15. Crea un vector (arreglo) de longitud 10 con valores equitativamente distribuidos entre 5 y 50.
a = np.linspace(5, 50, 10)

print(a)
```

```python
#Ejercicio 16. Crea un vector con valores de 0 a 20, y cambia el signo de los números de 9 a 15.
a = np.arange(0, 21)

a[9:16] = -a[9:16]

print(a)
```

```python
#Ejercicio 17. Convierte el arreglo `a = np.array([1, 2, 3])` a una lista de Python.
a = np.array([1, 2, 3])

lista = a.tolist()

print(lista)
```

```python
#Ejercicio 18. Convierte la lista `l = [1, 2, 3]` a un arreglo de Numpy.
l = [1, 2, 3]

a = np.array(l)

print(a)
```

```python
#Ejercicio 19. Dada la matriz `m = np.array([[1, 2], [3, 4]])`, verifica qué números son pares y almacena los resultados en otra matriz.
m = np.array([[1, 2],
              [3, 4]])

pares = (m % 2 == 0)

print(pares)
```

```python
# ejercicio 20. Dado el arreglo `m = np.array([[1, 2, 3], [4, 5, 6]])`, encuentra el número de filas y columnas de la matriz.
m = np.array([[1, 2, 3],
              [4, 5, 6]])

filas, columnas = m.shape

print("Filas:", filas)
print("Columnas:", columnas)
```
